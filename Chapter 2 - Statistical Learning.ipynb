```python
flexibility = [0, 1, 2, 3, 4, 5]
bias = np.array([6, 4, 2, 1, .5, .1])
variance = [.1, .5, 1, 2, 4, 6]
training_error = bias * 1.1
test_error = [7, 5, 2, 2, 5, 7]
irreducible_error = [1.5, 1.5, 1.5, 1.5, 1.5, 1.5]
```

```python
plt.plot(flexibility, bias,
         flexibility, variance,
         flexibility, training_error,
         flexibility, test_error,
         flexibility, irreducible_error)
plt.xlabel('Flexibility')
plt.legend(labels=('Bias', 'Variance', 'Training Error', 'Test Error', 'Irreducible Error'))
plt.show()
```

Bias decreases with more flexible models, as an inflexible model is unlikely to match the real-life situation

Variance increases with more flexible models, since noise can affect the model more

Training error decreases with more flexible models, and can even decrease to 0 if we overfit to all of the error

Test error is optimized somewhere in the middle - too inflexible and the model will be too far from the truth of the data set and too flexible can cause over-fitting

The irreducible error is constant noise inherent in the data. When the training error is less than the irreducible error, there has been overfitting. The test error is bounded below by the irreducible error.

## Question 5

What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

A flexible method is preferred when the relationship appears very complex and non-linear, as it has a lower bias. The disadvantages are that it can follow the noise too closely and overfit the data, so it's not good in situations where you have relatively few samples. A flexible method is also harder to interpret the results, so it's better when you're interested in prediction and not inference. 

A less flexible method is preferred when you want to be able to easily interpret the results for inferrence. It will have a higher bias, but less variance so it's more robust to noise. Less flexible is preferred when you have a smaller data set and a large number of features relative to the number of observations. 

## Question 6

Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification as opposed to a non-parametric approach? What are the disadvantages?

The parametric approach is less flexible - it assumes a form for $f$ and locks you into a model with a certain structure. With the parametric approach, you estimate the parameters for the assumed form of $f$. Advantages are its simplicity, direct inference of the predictors, and less computational time to optimize the parameters for the model. The disadvantage is the lack of flexibility. If you choose a poor parametric model, it may not match real-life very well and cause a significant error from bias

The non-parametric approach does not assume a form for $f$ so it's highly flexible. However, it requires a very large number of observations in order to accurately estimate f, and it's possible to overfit to the observations.
